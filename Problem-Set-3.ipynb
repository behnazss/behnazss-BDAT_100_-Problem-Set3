```python
import pandas as pd

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, delimiter='|')

mean_age_by_occupation = users.groupby('occupation')['age'].mean()
print("Mean Age by Occupation:")
print(mean_age_by_occupation)

male_ratio_by_occupation = users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').mean()).sort_values(ascending=False)
print("\nMale Ratio by Occupation:")
print(male_ratio_by_occupation)

min_max_age_by_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print("\nMin and Max Age by Occupation:")
print(min_max_age_by_occupation)

mean_age_by_occupation_and_sex = users.groupby(['occupation', 'gender'])['age'].mean()
print("\nMean Age by Occupation and Gender:")
print(mean_age_by_occupation_and_sex)

total_by_occupation = users.groupby('occupation').size()
percentage_by_occupation = users.groupby(['occupation', 'gender']).size() / total_by_occupation * 100
print("\nPercentage by Occupation and Gender:")
print(percentage_by_occupation)
```

```python
import pandas as pd
import numpy as np

# Step 2. Import the dataset from the attached file 'wind.txt':
data = pd.read_csv(r'C:\Users\behna\OneDrive\Desktop\Problem set3\wind.txt', delimiter='\s+', parse_dates=[[0, 1, 2]])
print(data.head())

print(data.columns)

# Step 3. Assign it to a variable called `data` and replace the first 3 columns by a proper datetime index:
data['date'] = pd.to_datetime(data['Yr_Mo_Dy'])
data = data.set_index('date')
data = data.drop(['Yr_Mo_Dy'], axis=1)
# Step 4. Create a function to fix the year and apply it:
def fix_year(year):
    if year > 1989:
        return year - 100
    else:
        return year

data.index = data.index.map(lambda x: x.replace(year=fix_year(x.year)))

# Step 5. Set the right dates as the index and change the data type to `datetime64[ns]`:
data.index = pd.to_datetime(data.index, format='%y-%m-%d')
#Step 6. Compute how many values are missing for each location over the entire record:
missing_values = data.isnull().sum()
print("\nmissing_values: ", missing_values)
#Step 7. Compute how many non-missing values there are in total:
non_missing_values = data.notnull().sum().sum()
print("\nnon_missing_values: ", non_missing_values)
mean_windspeed = data.mean().mean()
print("\nmean_windspeed:", mean_windspeed)

loc_stats = pd.DataFrame()
loc_stats['min'] = data.min()
loc_stats['max'] = data.max()
loc_stats['mean'] = data.mean()
loc_stats['std'] = data.std()
print("\nloc_stats:", loc_stats)

day_stats = pd.DataFrame()
day_stats['min'] = data.min(axis=1)
day_stats['max'] = data.max(axis=1)
day_stats['mean'] = data.mean(axis=1)
day_stats['std'] = data.std(axis=1)
print("\nday_stats:", day_stats)

jan_avg = data[data.index.month == 1].mean()
print("\njan_avg:", jan_avg)
yearly_data = data.resample('Y').mean()
print("\nyearly_data:", yearly_data)

monthly_data = data.resample('M').mean()
print("\nmonthly_data:", monthly_data)
weekly_data = data.resample('W').mean()
print("\nweekly_data:", weekly_data)
weekly_stats = pd.DataFrame()
weekly_stats['min'] = data['1961-01-02':'1961-12-31'].resample('W').min().min(axis=1)
weekly_stats['max'] = data['1961-01-02':'1961-12-31'].resample('W').max().max(axis=1)
weekly_stats['mean'] = data['1961-01-02':'1961-12-31'].resample('W').mean().mean(axis=1)
weekly_stats['std'] = data['1961-01-02':'1961-12-31'].resample('W').std().std(axis=1)
print("\nweekly_stats:", weekly_stats)

```

Question 6 Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("C:\\Users\\behna\\OneDrive\\Desktop\\Problem set3\\us-marriages-divorces-1867-2014.csv")

# Set the 'Year' column as the index
df.set_index('Year', inplace=True)

# Create the line plot
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot the number of marriages per capita
plt.plot(df.index, df['Marriages_per_1000'], label='Marriages per 1000', color='yellowgreen', marker='o')  # Green-Olive hex code

# Plot the number of divorces per capita
plt.plot(df.index, df['Divorces_per_1000'], label='Divorces per 1000', color='#01796F', marker='o')  # Pine hex code

# Set the title and labels for the axes
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.xlabel('Year')
plt.ylabel('Number per 1000 people')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("C:\\Users\\behna\\OneDrive\\Desktop\\Problem set3\\us-marriages-divorces-1867-2014.csv")

# Filter the data for the years 1900, 1950, and 2000
years = [1900, 1950, 2000]
df_filtered = df[df['Year'].isin(years)]

# Create the vertical bar chart
plt.figure(figsize=(8,5))  # Set the figure size

# Set the width of the bars
bar_width =5.80

# Position of the bars on the x-axis
marriages_pos = df_filtered.index
divorces_pos = df_filtered.index + bar_width

# Plot the number of marriages per capita
plt.bar(marriages_pos, df_filtered['Marriages_per_1000'], bar_width, label='Marriages per 1000', color='#01796F')

# Plot the number of divorces per capita
plt.bar(divorces_pos, df_filtered['Divorces_per_1000'], bar_width, label='Divorces per 1000', color='yellowgreen')

# Set the title and labels for the axes
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1900, 1950, and 2000)')
plt.xlabel('Year')
plt.ylabel('Number per 1000 people')

# Set the x-axis tick positions and labels
plt.xticks(df_filtered.index + bar_width/2, years)

# Show the legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

Question 8 Create a horizontal bar chart that compares the deadliest actors in Hollywood.
Sort the actors by their kill count and label each bar with the corresponding actor's name.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("C:\\Users\\behna\\OneDrive\\Desktop\\Problem set3\\actor_kill_counts.csv")

# Sort the actors by their kill count in descending order
df_sorted = df.sort_values(by='Count', ascending=False)

# Create the horizontal bar chart
plt.figure(figsize=(10, 5))  # Set the figure size

# Plot the horizontal bars
plt.barh(df_sorted['Actor'], df_sorted['Count'], color='#01796F')

# Label each bar with the corresponding actor's name and kill count
for index, value in enumerate(df_sorted['Count']):
    plt.text(value, index, str(value), ha='left', va='center', color='black', fontsize=10)

# Set the title and labels for the axes
plt.title('Deadliest Actors in Hollywood')
plt.xlabel('Kill Count')
plt.ylabel('Actor')

# Show the plot
plt.tight_layout()
plt.show()
```

Question 9 Create a pie chart showing the fraction of all Roman Emperors that were assassinated.  Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("C:\\Users\\behna\\OneDrive\\Desktop\\Problem set3\\roman-emperor-reigns.csv")

# Count the number of emperors who were assassinated
assassinated_emperors = df[df['Cause_of_Death'] == 'Assassinated'].shape[0]

# Count the number of emperors who were not assassinated
non_assassinated_emperors = df[df['Cause_of_Death'] != 'Assassinated'].shape[0]

# Create the pie chart
plt.figure(figsize=(5, 5))  # Set the figure size

# Data to plot
labels = ['Assassinated', 'Non-Assassinated']
sizes = [assassinated_emperors, non_assassinated_emperors]

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['yellowgreen', '#01796F'])

# Set the title
plt.title('Fraction of Roman Emperors Assassinated')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show the pie chart
plt.show()
```

Question 10 Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009.  Don't forget to label your axes!  Color each dot according to its year. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("C:\\Users\\behna\\OneDrive\\Desktop\\Problem set3\\arcade-revenue-vs-cs-doctorates.csv")

# Extract data for the scatter plot
arcade_revenue = df['Total Arcade Revenue (billions)']
cs_phds_awarded = df['Computer Science Doctorates Awarded (US)']
years = df['Year']

# Create the scatter plot
plt.figure(figsize=(10, 6))  # Set the figure size

# Data to plot
for year in years.unique():
    plt.scatter(arcade_revenue[years == year], cs_phds_awarded[years == year], label=year)

# Set labels and title
plt.xlabel('Total Revenue Earned by Arcades (billions)')
plt.ylabel('Number of Computer Science PhDs Awarded (US)')
plt.title('Relationship between Arcade Revenue and Computer Science PhDs (2000-2009)')

# Add legend
plt.legend(title='Year')

# Show the scatter plot
plt.show()
```
